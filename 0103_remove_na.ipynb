```python
import pandas as pd
```

```python
def remove_na():
    PATH = './data_reshape.xlsx'
    OUTPATH = './data_without_na_01.xlsx'
    df = pd.read_excel(PATH)
    print(df.shape)
    print(df.isnull().sum())
    
    df.dropna(subset=['术前诊断','实施手术'],inplace=True)
    
    avg_weight = df['体重'].mean().round(0)
    print('平均体重:',avg_weight)
    df['体重'].fillna(value=avg_weight,inplace=True)
    
    avg_height = df['身高'].mean().round(0)
    print('平均身高:',avg_height)
    df['身高'].fillna(value=avg_height,inplace=True)
    df.dropna(subset=['性别','年龄'],inplace=True)
    print(df.shape)
    print(df.isnull().sum())
    
    # 处理负数
    df.loc[df['身高'] <= 0,'身高'] = avg_height
    df.loc[df['体重'] <= 0,'体重'] = avg_weight
    df.drop(df[df['年龄'] > 100].index,inplace=True)
    df.to_excel(OUTPATH)
```

```python
remove_na()
```
